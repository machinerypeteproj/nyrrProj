```python
results = dict()
for _, row in full_women_results.iterrows():
    if row["FinishPlace"] == 1:
        if results.get(row["Name"], None):
            results[row["Name"]]["TopFives"] += 1
        else:
            results[row["Name"]] = {"Name": row["Name"], "TopFives": 1}
```

```python
results = [results[key] for key in results.keys()]
```

```python
results = pd.DataFrame(results)
```

```python
results.sort_values("TopFives", ascending=False)
```
